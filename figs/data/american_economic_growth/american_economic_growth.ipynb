```python
import pandas as pd
import sys

sys.path.append('../_styling/')
from style import prep_plot, GRAY1, GRAY2
```

```python
df = pd.read_csv("original_fig16-5.csv")
df.head()
```

```python
plt = prep_plot()

colors = [GRAY2 if x < 1 else GRAY1 for x in df['Growth']]

plt.bar(x=df['Decade'], height=df['Growth'], color=colors, width=8)
plt.xticks(df['Decade'])
plt.xticks(rotation = 0)
plt.xlabel("")

plt.text(1950, 3.8, "Golden Age", weight='bold', ha='center', color=GRAY1)
plt.plot([1926, 1974], [3.6, 3.6], lw=2, marker='|', color=GRAY1)

plt.text(1995, 1.2, "Digital Stagnation", weight='bold', ha='center', color=GRAY2)
plt.plot([1976, 2014], [1, 1], lw=2, marker='|', color=GRAY2)

plt.ylabel("Percent\nper year", rotation=0, labelpad=35, weight='bold')
plt.savefig("golden.png", bbox_inches="tight")
```
